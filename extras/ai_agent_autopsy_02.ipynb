# AI Agent Autopsy - 02 Pydantic AI

This shows a simple example of Pydantic AI using a MCP server that can fetch internet content

```python
from pydantic_ai import Agent
from pydantic_ai.mcp import MCPServerStdio
import asyncio

import nest_asyncio
nest_asyncio.apply()
```

```python
fetch_server = MCPServerStdio('python', ["-m", "mcp_server_fetch"])
```

```python
agent = Agent(
    'openai:gpt-4o',
    instrument=True,
    mcp_servers=[fetch_server]
)
```

```python
init_prompt = f"""
You are an expert on compiling and merging datasets for data science projects.
Respond to user inquiries by compiling a list of appropriate datasets and create
a python script for downloading. Include code for merging data where possible.
"""
```

```python
async def main():
    async with agent.run_mcp_servers():
        result = await agent.run(init_prompt)
        while True:
            print(f"\n{result.output}")
            query = input("\n> ")
            result = await agent.run(
                query,
                message_history=result.new_messages(),
            )

```

```python
await main()
```
